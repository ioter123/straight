```python

#pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from konlpy.tag import Twitter
from collections import Counter
import pandas as pd
#사전 실행
twitter = Twitter()

data = pd.read_csv('피갤 잔처리.csv',engine='python', encoding= 'utf-8')
#단어 리스트 만들기
word_list = []
word_list = data['제목']

#형태소 분리
sentences_tag = []

for sentence in word_list:
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
print(morph)

#명사만 추출
noun_list = []
for sentence in sentences_tag:
    for word, tag in sentence:
        if tag in['Noun']:
            noun_list.append(word)
            
# 명사 중 두 글자 이상만 추출

noun_list2 = [n for n in noun_list if len(n) > 1]
            
# 두 글자 이상인 명사 개수 카운트

counts = Counter(noun_list2)
tags2 = counts.most_common()
tags2

#워드클라우드 만들기

wordcloud = WordCloud(font_path='C:\Windows\Fonts\malgun.ttf', background_color='white', width=800,height=600)

print(dict(tags2))

cloud = wordcloud.generate_from_frequencies(dict(tags2))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
