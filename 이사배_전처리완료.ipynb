```python
import pandas as pd
import re
data = pd.read_csv('이사배_날것완성.csv',engine='python',encoding='utf8')
data
```

```python
import re
#이모티콘 제거
emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"
                           u"\U0001F300-\U0001F5FF"
                           u"\U0001F680-\U0001F6FF"
                           u"\U0001F1E0-\U0001F1FF"
                           u"\U00002702-\U000027B0" #이거하니까 다른 물음표도없어짐
                           "]+", flags=re.UNICODE)

#https://www.it-swarm.dev/ko/python/python%EC%9D%98-%EB%AC%B8%EC%9E%90%EC%97%B4%EC%97%90%EC%84%9C-%EC%9D%B4%EB%AA%A8%ED%8B%B0%EC%BD%98-%EC%A0%9C%EA%B1%B0/1057057628/

#분석에 어긋나는 불용어구 제외 (특수문자, 의성어)
han = re.compile(r'[ㄱ-ㅎㅏ-ㅣ!?~,".\n\r#\ufeff\u200d]')
```

```python
comment_list = []
for i in range(len(data)):
    comment_list.append(data['제목'].iloc[i])
    
comment_result = []

for i in comment_list:
    tokens = re.sub(emoji_pattern,"",i)
    tokens = re.sub(han,"",tokens)
    comment_result.append(tokens)

comment_result = pd.DataFrame(comment_result, columns=["제목"])
```

```python
# 해도 다 안걸러짐..하나하나 다시 해봄
data.제목 = comment_result
data['제목'] = data.제목.replace('\!', '', regex=True)
data['제목'] = data.제목.replace('\?', '', regex=True)
data['제목'] = data.제목.replace('\,', '', regex=True)
data['제목'] = data.제목.replace('\'', '', regex=True)
data['제목'] = data.제목.replace('\.', '', regex=True)
data['제목'] = data.제목.replace('\+', '', regex=True)
data['제목'] = data.제목.replace('\&', '', regex=True)
data['제목'] = data.제목.replace('\/', '', regex=True)
data['제목'] = data.제목.replace('\#', '', regex=True)
data['제목'] = data.제목.replace('\[Eng]', '', regex=True)
data
```

```python
like_ls = []
view_ls = []
unlike_ls = []
comment_ls = []
date_ls=[]

for i in range(len(data)):
    if '천' in data['좋아요'].iloc[i]:
        a = ''.join(re.findall('[0-9]',data['좋아요'].iloc[i]))
        if len(a) == 2:
            c = a+'00'
        else:
            c = a+'000'
    elif '만' in data['좋아요'].iloc[i]:
        b = ''.join(re.findall('[0-9]',data['좋아요'].iloc[i]))
        if len(b) ==2:
            c = b+'000'
        else:
            c = b+'0000'
    else:
        c = data['좋아요'].iloc[i]
    like_ls.append(c)
    
    if '천' in data['싫어요'].iloc[i]:
        aa = ''.join(re.findall('[0-9]',data['싫어요'].iloc[i]))
        if len(a) == 2:
            cc = aa+'00'
        else:
            cc = aa+'000'
    elif '만' in data['싫어요'].iloc[i]:
        bb = ''.join(re.findall('[0-9]',data['싫어요'].iloc[i]))
        if len(bb) ==2:
            cc = bb+'000'
        else:
            cc = bb + '0000'
    else:
        cc = data['싫어요'].iloc[i]
    unlike_ls.append(cc)
    
    view0 = ''.join(re.findall('[0-9]',data['조회수'].iloc[i]))
    view_ls.append(view0)
    
    comment0 = ''.join(re.findall('[0-9]',data['댓글수'].iloc[i]))
    comment_ls.append(comment0)
    
    date0 = ''.join(re.findall('[.0-9]',data['영상게시일'].iloc[i]))
    date_ls.append(date0[:-1])
    

data['조회수'] = view_ls
data['영상게시일']=date_ls
data['댓글수'] = comment_ls
data['좋아요'] = like_ls
data['싫어요'] = unlike_ls
```

```python
data['영상게시일'] = pd.to_datetime(data.영상게시일)
```

```python
import numpy as np
video_time_list = np.array(data['영상길이'].tolist())
```

```python
def stime(text):
    time = text.split(':')
    if len(time) == 1:
        s = [int(i) for i in text.split(':')]
        return s
    elif len(time) == 2:
        m, s = [int(i) for i in text.split(':')]
        return 60*m + s
    else:
        h, m, s = [int(i) for i in text.split(':')]
        return 3600*h + 60*m + s
    
video_time_seperate_list = []

for time in video_time_list:
    video_time_seperate_list.append(stime(time))
```

```python
data.영상길이 = video_time_seperate_list
```
