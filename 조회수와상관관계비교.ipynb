```python
import pandas as pd
```

```python
# csv 파일 불러오기
data = pd.read_csv('temp/final2.csv')
```

```python
data.info()
```

```python
data
```

```python
# 좋아요 싫어요 '.' 없애고 0붙이고 천,만 없애고 '0'붙이고 숫자타입으로 변환

data['좋아요'] = data.좋아요.replace('천', '0', regex=True)
data['좋아요'] = data.좋아요.replace('만', '00', regex=True)
for i in range(100):
    if '.' in data.좋아요[i]:
        data['좋아요'][i] = data.좋아요[i].replace('.', '',)
        data['좋아요'][i] = data.좋아요[i] + '0'

data['싫어요'] = data.싫어요.replace('천', '0', regex=True)
data['싫어요'] = data.싫어요.replace('만', '00', regex=True)
for i in range(100):
    if '.' in data.싫어요[i]:
        data['싫어요'][i] = data.싫어요[i].replace('.', '',)
        data['싫어요'][i] = data.싫어요[i] + '0'
        
data['좋아요'] = pd.to_numeric(data.좋아요)
data['싫어요'] = pd.to_numeric(data.싫어요)

data.info()
```

```python
# 게시날짜 데이터타입으로
data['영상게시일'] = pd.to_datetime(data.영상게시일)
```

```python
data.info()
```

```python
# 제목에서 특수문자 지우기
data['제목'] = data.제목.replace('\!', '', regex=True)
data['제목'] = data.제목.replace('\?', '', regex=True)
data['제목'] = data.제목.replace('\,', '', regex=True)
data['제목'] = data.제목.replace('\'', '', regex=True)
data['제목'] = data.제목.replace('\.', '', regex=True)
data['제목'] = data.제목.replace('\+', '', regex=True)
data['제목'] = data.제목.replace('\&', '', regex=True)
data['제목'] = data.제목.replace('\/', '', regex=True)
data['제목'] = data.제목.replace('\#', '', regex=True)
data['제목'] = data.제목.replace('\%', '', regex=True)
data.제목
```

```python
data
```

```python
# 영상 길이 초로 변환
data['temp1'] = None
data['temp2'] = None
for i in range(len(data.좋아요)):
    data['temp1'][i] = data.영상길이[i][0 : data.영상길이[i].find(':')]
    data['temp2'][i] = data.영상길이[i][data.영상길이[i].find(':')+1 : len(data.영상길이[i])]
data['temp1'] = pd.to_numeric(data.temp1)
data['temp2'] = pd.to_numeric(data.temp2)
data['영상길이'] = (data.temp1 * 60) + data.temp2
data.drop(['temp1', 'temp2'], axis=1, inplace=True)
data
```
